##### **** Configure the transform parameters. The set of dictionary keys holding ProglangSelect configuration for values are as follows: 
* proglang_select_allowed_langs_file - specifies path to allowed languages file.
* proglang_select_language_column - specifies column name that contains programming language. By default, "language" is used.
#####

##### ***** Import required classes and modules

````python
from dpk_proglang_select.runtime import ProglangSelect
````

##### ***** Setup runtime parameters and invoke the transform

````python
ProglangSelect(
    input_folder= "./test-data/input",
    output_folder= "./output",
    proglang_select_output_column='lang_selected',
    proglang_select_language_column='language',
    proglang_select_allowed_langs_file="./test-data/languages/allowed-code-languages.txt"
).transform()
````
