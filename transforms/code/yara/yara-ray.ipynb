##### ***** Import required classes and modules

````python
from dpk_yara.ray.runtime import Yara
import pandas as pd
````

##### ***** Generate a small sample input (benign row + EICAR test row) if `test-data/input/` is empty. The YARA transform scans `binary_contents` bytes column.

````python
import os, subprocess, sys
if not os.path.exists('test-data/input/sample.parquet'):
    subprocess.run([sys.executable, 'scripts/make_test_data.py'], check=True)
````

##### ***** Setup runtime parameters for this transform and invoke the transform. The set of all possible parameters and their default values are [here](./README.md).

`yara_rules_dir` points at the committed `test-rules/` directory (EICAR test rule only) so this notebook runs hermetically without running `scripts/fetch_rules.py`. For production use, populate `rules/` via `make fetch-rules` (or set `DPK_YARA_RULES_DIR` / pass a different path).

````python
Yara(input_folder="test-data/input",
     output_folder="output",
     yara_input_column="binary_contents",
     yara_rules_dir="test-rules",
     run_locally=True,
     ).transform()
````
