##### ***** Setup access keys and Import required classes and modules

````python
import os, json
from dpk_ededup.transform_python import Ededup
os.environ['S3_ACCESS_KEY']=os.environ['COS_DPK_ACCESS_KEY']
os.environ['S3_SECRET_KEY']=os.environ['COS_DPK_SECRET_KEY']
os.environ['S3_ENDPOINT']= "https://s3.us-south.cloud-object-storage.appdomain.cloud"
````

##### ***** Setup runtime parameters for this transform and invoke the transform
* ededup_doc_column - specifies name of the column containing documents
* ededup_doc_id_column - specifies the name of the column containing a document id

````python
%%time
Ededup(input_folder="s3://bucket-wikimedia-1/take6_md",
    output_folder="s3://bucket-wikimedia-1/test-MT",
    ededup_doc_column="contents",
    ededup_doc_id_column="document_id").transform()
````
