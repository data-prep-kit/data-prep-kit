##### 1 • Import the transform runtime and utilities

````python
import pandas
from dpk_ml_filter.runtime import MLFilter
from dpk_ml_filter.transform import get_transform_params
````

##### 2 • Required parameters for the transform
| Name  | Default Value | Description |
|------------|----------|--------------|
| **lm_filter_config** | **_text_** | File name with the conditions for the filter, in YAML format. |
| **lm_filter_lang_column_name** | **lang** | The column name with language identifier used in the filter conditions. |
| **lm_filter_column_prefix** | _not set_ | A prefix for the names referenced in the conditions file.|
| **lm_filter_ignore_missing_columns** | **False** | By default the transform will fail if any for the values referenced in the conditions is missing. Set this parameter to ignore all the missing columns. |


##### 3 • Excerpt from a conditions file

```yaml
# the default section is applied to all languages, along with any language specific condition
default:
    avg_word_length_min: 3.2
    avg_word_length_max: 7.0

# the laguage sections define conditions for each language separately

en:
    num_words_min: 110
    lid_score_min: 0.5
fr:
    num_words_min: 100
    lid_score_min: 0.6

# by default a all enties with a language not specified here will be removed from the output, unless an <other> section is defined
pt: 

# the other section is applied to all the languages without an explicit section
other:
    num_words_min: 1000
```

##### 4 • Setup the transform

````python
transform = MLFilter(
    ml_filter_config="test-data/input/cleansing-config.yaml",
    ml_filter_lang_column_name="lang",
    ml_filter_column_prefix="e_",
    input_folder="test-data/input", 
    output_folder="output" 
)
````

##### 5 • Run

````python
transform.transform()
````

##### 6 • Show the number of rows before and after applying the transform

````python
before = len(pandas.read_parquet('test-data/input/1.parquet', engine='pyarrow'))
after = len(pandas.read_parquet('output/1.parquet', engine='pyarrow'))
print(f"before: {before}, after: {after}")
````
