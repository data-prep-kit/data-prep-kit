##### **** Setup runtime parameters and invoke the transform. For a full list of parameters for the filter transform, please see [here](./README.md).

````python
Filter(input_folder= "test-data/ds01/input/parquet",
        output_folder= "output/parquet",
        filter_columns_to_drop= ["extra", "cluster"],
        filter_criteria_list= [
            "docq_total_words > 100 AND docq_total_words < 400",
            "ibmkenlm_docq_perplex_score < 250",
        ],
        filter_input_arrow_folder= "test-data/ds01/input/arrow",
        filter_output_arrow_folder = "output/arrow",
        filter_doc_id_column_name = "document_id",
        filter_logical_operator= "AND").transform()
````
